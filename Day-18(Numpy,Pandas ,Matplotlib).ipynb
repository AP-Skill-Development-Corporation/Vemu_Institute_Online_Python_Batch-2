```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [5, 10, 2, 18, 6, 15, 8]

plt.plot(x, y)
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.title('Line Plot')
plt.show()
```

```python
help(plt.plot)
```

```python
import numpy as np

x = np.radians(np.arange(-360, 361, 15))
siny = np.sin(x)
cosy = np.cos(x)

plt.plot(x, siny, linestyle = '-.', linewidth = 1.5, color = 'g', marker = '*', label = 'Sin Plot')
plt.show()
plt.plot(x, cosy, linestyle = '--', linewidth = 1.5, color = 'r', marker = 'o', label = 'Cos Plot')
plt.legend()
plt.grid()
plt.show()
```

# Pie chart

```python
#plt.pie()
import matplotlib.pyplot as plt
labels=["python","c","java","c++","sql",".net"]
slices=[9000,3000,2400,1400,5000,2000]
explode=(1,0,0,0,0,0)
plt.pie(slices,labels=labels,wedgeprops={"edgecolor":"red"},startangle=90,explode=explode,shadow=True)
plt.tight_layout()
plt.show()
```
